```python
import StockMarketHelper as helper

# Filter symbols to get a list of companies
all_symbols = helper.filter_symbol(search='', return_all=True)
print(all_symbols)

# Load data for all companies
all_data = []
for symbol in all_symbols['key']:
    data = helper.read_from_key(symbol)
    data['symbol'] = symbol
    all_data.append(data)

# Concatenate all data into a single DataFrame
all_data_df = pd.concat(all_data)
print(all_data_df.head())
```

```python
# Feature Engineering (Example: MA50 and Volatility)
all_data_df['MA50'] = all_data_df.groupby('symbol')['close'].transform(lambda x: x.rolling(window=50).mean())
all_data_df['Volatility'] = all_data_df.groupby('symbol')['close'].transform(lambda x: x.rolling(window=50).std())

# Drop rows with NaN values
all_data_df.dropna(inplace=True)

# Scaling the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
all_data_df[['open', 'high', 'low', 'close', 'volume', 'MA50', 'Volatility']] = scaler.fit_transform(
    all_data_df[['open', 'high', 'low', 'close', 'volume', 'MA50', 'Volatility']])
```

```python
def create_sequences(data, seq_length):
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = data[i + seq_length, 3]  # predict the 'close' price
        yield np.array(x), np.array(y)
```

```python
def batch_generator(data, symbols, seq_length, batch_size):
    while True:
        X_batch = []
        y_batch = []
        for symbol in symbols:
            symbol_data = data[data['symbol'] == symbol].drop(columns=['symbol']).values
            seq_gen = create_sequences(symbol_data, seq_length)
            for x, y in seq_gen:
                X_batch.append(x)
                y_batch.append(y)
                if len(X_batch) == batch_size:
                    yield np.array(X_batch), np.array(y_batch)
                    X_batch = []
                    y_batch = []
```

```python
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = data[i + seq_length, 3]  # predict the 'close' price
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 50
X = []
y = []

# Create sequences for each symbol
for symbol in all_data_df['symbol'].unique():
    symbol_data = all_data_df[all_data_df['symbol'] == symbol].drop(columns=['symbol']).values
    X_symbol, y_symbol = create_sequences(symbol_data, seq_length)
    X.append(X_symbol)
    y.append(y_symbol)

X = np.concatenate(X)
y = np.concatenate(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True, input_shape=(seq_length, X_train.shape[2])),
    tf.keras.layers.LSTM(64, return_sequences=False),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```
